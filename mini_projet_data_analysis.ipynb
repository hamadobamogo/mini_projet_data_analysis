#### Importation des données

```python
import pandas as pd

# Ouverture du fichier texte et lecture de son contenu
with open('BF.txt', 'r',encoding='utf-8') as fichier:
    lignes = fichier.readlines()

# Suppression des sauts de lignes 
donnees = [ligne.strip().split('\t') for ligne in lignes]  

# Création d'un DataFrame à partir des données
df = pd.DataFrame(donnees)

# Affichage du DataFrame
df.head()
```

### 3°) Appliquer les opérations de prétraitement et filtres nécessaires à ce fichier, pour ne garder que les colonnes correspondantes

#### a°) Identifiants, Noms de lieux, latitudes, longitudes

#### Ces colonnes Identifiants, Noms de lieux, latitudes, longitudes ont pour indice respectivement 0;1;4 et 5

```python
# Conservation des colonnes:  Identifiants,Noms de lieux, latitudes et longitudes
L = [0,1,4,5] # Liste des colonnes concernées
data = df.iloc[:,L]
data
```

#### b°) Renommez les avec les noms suivants : 'ID', 'location_name', 'lat', 'long'

```python
# Renommons les colonnes 0;1;4 et 5 par les noms suivants respectivement 'ID'; 'location_name'; 'lat' et 'long'
data = data.rename(columns={0: 'Id', 1: 'location_name', 4: 'lat', 5: 'long'})
data
```

#### c°) Sauvegarder ces données dans un fichier CSV, nommez-le burkina_location.csv

```python
# Sauvegardons ces données dans un fichier CSV nommé burkina_location.csv
data.to_csv('burkina_location.csv', index=False)
```

### 4°) Opérations sur le fichier CSV burkina_location.csv

```python
# importation des données
BF_data = pd.read_csv('burkina_location.csv')
BF_data
```

#### 4.1°) extraire les données contenant le nom 'gounghin', enregistrez-le sous le fichier gounghin.csv

```python
# extraction des données contenant le nom 'gounghin'
data_gounghin = BF_data[BF_data['location_name'] == 'Gounghin']
data_gounghin
```

```python
# Sauvegarde du résultat sous format csv
data_gounghin.to_csv('gounghin.csv', index=False)
```

#### 4.2°) extraire la sous-partie de la base de données (fichier burkina_location.csv), dont les noms les premières lettres des noms de lieux sont compris entre 'A' et 'P' (ordre alphabétique)

```python
# sélection des noms de lieux compris entre 'A' et 'P' et trions par ordre alphabétique
filtre = BF_data['location_name'].str[0].between('A','P')
lieux_entre_AP = BF_data[filtre].sort_values(by='location_name')
lieux_entre_AP
```

#### 4.3°) Identifiez respectivement, la latitude, la longitude minimale et les noms de lieux correspondants

```python
# Recherche de la latitude minimale et le nom du lieu correspondant
# trions le dataframe par ordre croissant suivant la latitude d'abord et recupérons le name de la 1ère ligne
indice = BF_data.sort_values(by = 'lat').iloc[0].name
indice
```

```python
# Affichons le nom du lieu et la latitude minimale
latitude_minimale = BF_data.loc[indice,['location_name','lat']]
latitude_minimale
```

```python
# Recherche de la longitude minimale et le nom du lieu correspondant
# trions le dataframe par ordre croissant suivant la longitude d'abord et recupérons le name de la 1ère ligne
indice = BF_data.sort_values(by = 'long').iloc[0].name
indice
```

```python
# Affichons le nom du lieu et la latitude minimale
latitude_minimale = BF_data.loc[indice,['location_name','long']]
latitude_minimale
```

#### 4.4°) Quels sont les lieux dont les coordonnées sont comprises entre (lat >= 11 et long <= 0.5)

```python
# Sélectionnons les lieux dont les coordonnées sont comprises entre (lat >= 11 et long <= 0.5)
filtre2 = (BF_data['lat'] >= 11) & (BF_data['long'] <= 0.5)
lieux_filtres = BF_data[filtre2]
lieux_filtres
```

```python
# Affichons les noms de ces lieux
lieux_filtres['location_name']
```
